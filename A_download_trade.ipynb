```python
import httpx
import asyncio
import pandas as pd

from os import path, mkdir
from toolkit.exchanges import Binance, Coinalyze
```

```python
binance = Binance()

markets = {
    "Binance": ["BTCUSDT", "ETHUSDT"]
}
coinalyze_markets = ["BTCUSD_PERP.A", "ETHUSD_PERP.A"]
```

```python
trade_data = []
print("Downloading trade data...")
async with httpx.AsyncClient() as client:
    for symbol in markets["Binance"]:
        trade_data.append(binance.trades(client, symbol, limit=1000))
    trade_data = await asyncio.gather(*trade_data)
    trade_data = dict(zip(markets["Binance"], trade_data))
print("Finished downloading.")
```

```python
try:
    coinalyze = Coinalyze()
    spot = coinalyze.spot_markets()
    futures = coinalyze.futures_markets()

    oi = coinalyze.open_interest(coinalyze_markets)
    fu = coinalyze.funding(coinalyze_markets)
except KeyError as err:
    print(err)
```

```python
df = {}
for symbol, trades in trade_data.items():
    df[symbol] = pd.DataFrame.from_dict(trades)
```
